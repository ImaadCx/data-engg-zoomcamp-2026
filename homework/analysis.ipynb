```python
import pandas as pd
```

```python
df_trips = pd.read_parquet ('green_tripdata_2025-11.parquet')

df_trips.dtypes
```

```python
 trips_short = df_trips[df_trips['trip_distance'] <= 1]
 len(trips_short)
```

```python
# Filter trips with distance less than 100 miles
filtered_trips = df_trips[df_trips['trip_distance'] < 100]

# Find the trip with the longest distance
longest_trip = filtered_trips.loc[filtered_trips['trip_distance'].idxmax()]

# Extract the pickup day
pickup_day = longest_trip['lpep_pickup_datetime'].day

print(f"Pickup day with longest trip distance: {pickup_day}")
print(f"Trip distance: {longest_trip['trip_distance']} miles")
```

```python
zones = pd.read_csv('taxi_zone_lookup.csv')
```
